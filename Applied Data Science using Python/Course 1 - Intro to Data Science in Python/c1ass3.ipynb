```python
def answer_seven():
    census_df = pd.read_csv('census.csv',index_col=6)
    counties_df = census_df[census_df['SUMLEV'] == 50]
    columns = ['POPESTIMATE2010',
                       'POPESTIMATE2011',
                       'POPESTIMATE2012',
                       'POPESTIMATE2013',
                       'POPESTIMATE2014',
                       'POPESTIMATE2015']
    max_value = counties_df[columns].max(axis = 1)
    max_value
    min_value = counties_df[columns].min(axis = 1)
    min_value
    diff=max_value-min_value
    return(diff.idxmax())
```
