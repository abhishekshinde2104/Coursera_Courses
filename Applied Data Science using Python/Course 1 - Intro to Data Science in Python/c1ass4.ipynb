```python
state_town = []
for line in data:
    if line[-6:] == '[edit]':
        state = line[:-6]
    elif '(' in line:
        town = line[:(line.index('(')-1)]
        state_town.append([state,town])
    else:
        town = line.rstrip()#rstrip() removes any trailing characters(characters as end of string) 
        #space is default trailing character to remove
        state_town.append([state,town])
```
